```python
import pandas as pd
```

```python
df= pd.read_csv('datasets/Admission_Predict.csv')
df.head(10)
```

```python
# We actully use the function .isnull() to create a boolean mask
mask=df.isnull()
mask.head(10)
```

```python
# We can drop all of those rows which have any missing data
df.dropna().head(15)
```

```python
# One of the handy functions that pandas has for working with missing value is the filling function fillna()
df.fillna(0,inplace=True)
df.head(10)
```

```python
df= pd.read_csv('datasets/log.csv')
df.head(20)
```

```python
# Next up is method parameter(). The two common fill values are ffill and bfill. ffill is for forward
# filling and it updates and an value for perticular cell with the value from the previous row. bfill is
# backward filling, which is opposite of ffill.
# In Pandas we can sort either by Index or by values
df= df.set_index('time')
df= df.sort_index()
df.head(20)
```

```python
# we'll notice that the index isn't really unique. Two users seem to be able to use the system at the same
# time. Use multi-level indexing on time and user together insted.

df= df.reset_index()
df= df.set_index(['time','user'])
df.head(10)
```

```python
# we can fill missing values using ffill.

df= df.fillna(method='ffill')
df.head()
```

```python
# We can also do customized fill-in with replace(). It allows to replacement from
# value-to-value, list, dictionary, regex. 

df= pd.DataFrame({'A':[1,1,2,3,4],
                 'B':[5,6,3,7,8],
                 'c':['a','b','c','d','e']})
df
```

```python
df.replace(1,100)
```

```python
df.replace([1,3],[100,300])
```
